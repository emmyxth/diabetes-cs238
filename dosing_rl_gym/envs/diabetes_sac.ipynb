```python
from diabetic_e import Diabetic0Env
import gym
import numpy as np
import random
from stable_baselines3 import SAC
from stable_baselines3.common.evaluation import evaluate_policy
```

```python
diabetic_sim = Diabetic0Env()
```

```python
diabetic_sim.set_episode_length(10)
```

```python
model = SAC("MlpPolicy", diabetic_sim, verbose=1)
```

```python
model.learn(total_timesteps=100000, log_interval=4)
model.save("SAC_diabetes")
```

```python
del model
```

```python
model = SAC.load("SAC_diabetes")
```

```python
mean_reward, std_reward = evaluate_policy(model, diabetic_sim, n_eval_episodes=10)
print(f"Mean reward: {mean_reward} +/- {std_reward}")
```
